# Task-4: Conversational AI with RAG

## 1. Problem Statement & Objective
The objective of this task is to develop a robust Conversational AI application using Retrieval-Augmented Generation (RAG). Instead of relying purely on a model’s parametric memory, the chatbot integrates an external, domain-specific knowledge base to retrieve context and generate more accurate, fact-based answers while maintaining conversational context over time. This approach ensures high relevance and mitigates hallucination issues that are common in generative models.


## 2. Dataset Loading & Preprocessing
For our custom corpus, we use a curated list of Artificial Intelligence and Machine Learning Wikipedia articles. The raw text from these pages is scraped, cleaned, and heavily chunked using LangChain's RecursiveCharacterTextSplitter. We used a chunk size of 1000 characters with an overlap of 150 to ensure semantic continuity. The resulting text segments (chunks) represent our structured dataset of knowledge.


## 3. Model Development & Training
Instead of 'training' a neural network from scratch, we employ an extensive RAG pipeline setup:
1. **Embedding**: The text chunks are passed through the `sentence-transformers/all-MiniLM-L6-v2` HuggingFace embedding model. These embeddings are mapped to high-dimensional vectors representing their semantic meaning.
2. **Vector Store**: A FAISS (Facebook AI Similarity Search) index is built as the vectorized document store. It acts as our highly efficient search engine.
3. **Generation**: We use the ChatGroq model `llama3-8b-8192` as our LLM for understanding user queries and synthesizing the final response. Additionally, a ConversationBufferMemory tracks conversational history context.


## 4. Evaluation with Relevant Metrics
Below we instantiate the core evaluation script utilizing LangChain. We'll test the relevance and accuracy of queries based on retrieval latency and visual feedback of response quality.


```python
import time
from chatbot import load_vector_store, build_chain, get_response

# Initialize setup
try:
    vector_store = load_vector_store()
    chain = build_chain(vector_store)
    print("Knowledge base and Retrieval Chain setup successfully.")
except Exception as e:
    print("Error:", e)
```

```python
# Evaluation queries
test_queries = [
    "What is Retrieval-Augmented Generation?",
    "Explain the concept of word embeddings in NLP.",
    "How are GANs trained?"
]

for q in test_queries:
    print(f"\n{'='*50}")
    print(f"Q: {q}")
    
    start_time = time.time()
    result = get_response(chain, q)
    end_time = time.time()
    
    print(f"\nAnswer: {result['answer']}")
    print(f"\nSources: {result['sources']}")
    print(f"Latency: {end_time - start_time:.4f} seconds")
```
