**Tools:** pandas, matplotlib, seaborn

## Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean visual style for all plots
sns.set_theme(style='whitegrid', palette='Set2')
plt.rcParams['figure.figsize'] = (8, 5)

print('Libraries imported successfully.')
```

## Step 2: Load the Dataset

```python
# Load the Iris dataset using seaborn (no file download needed)
df = sns.load_dataset('iris')

print('Dataset loaded successfully!')
print(f'Shape: {df.shape}')  # (rows, columns)
print(f'Columns: {list(df.columns)}')
```

## Step 3: Initial Inspection

```python
# First 5 rows
print('First 5 rows:')
df.head()
```

```python
# Data types and non-null counts
print('Dataset Info:')
df.info()
```

```python
# Summary statistics for all numeric columns
print('Descriptive Statistics:')
df.describe()
```

```python
# Check for missing values
print('Missing values per column:')
print(df.isnull().sum())

# Check class distribution
print('\nSpecies distribution:')
print(df['species'].value_counts())
```

## Step 4: Scatter Plot – Feature Relationships

We use a pairplot to see how each feature relates to the others, grouped by species.

```python
# Pairplot: scatter matrix showing all feature combinations by species
sns.pairplot(df, hue='species', diag_kind='kde', markers=['o', 's', 'D'])
plt.suptitle('Iris Dataset – Pairwise Feature Relationships', y=1.02, fontsize=14)
plt.tight_layout()
plt.savefig('scatter_pairplot.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved: scatter_pairplot.png')
```

```python
# Individual scatter: petal_length vs petal_width (best separator between species)
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', s=80)
plt.title('Petal Length vs Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.savefig('scatter_petal.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved: scatter_petal.png')
```

## Step 5: Histograms – Value Distributions

Histograms show how each feature is distributed across the full dataset.

```python
# 4 histograms, one for each numeric feature
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, feature in enumerate(features):
    sns.histplot(data=df, x=feature, hue='species', bins=20, kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {feature.replace("_", " ").title()}')
    axes[i].set_xlabel(feature.replace('_', ' ').title() + ' (cm)')

plt.suptitle('Feature Distributions by Species', fontsize=14)
plt.tight_layout()
plt.savefig('histograms.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved: histograms.png')
```

## Step 6: Box Plots – Outlier Detection

Box plots show the spread and highlight any outliers per feature.

```python
# 4 box plots for each feature, grouped by species
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, feature in enumerate(features):
    sns.boxplot(data=df, x='species', y=feature, ax=axes[i], palette='Set2')
    axes[i].set_title(f'Box Plot: {feature.replace("_", " ").title()}')
    axes[i].set_xlabel('Species')
    axes[i].set_ylabel(feature.replace('_', ' ').title() + ' (cm)')

plt.suptitle('Feature Outlier Analysis by Species', fontsize=14)
plt.tight_layout()
plt.savefig('boxplots.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved: boxplots.png')
```

## Step 7: Correlation Heatmap (Bonus)

A heatmap of the correlation matrix shows which features are strongly related to each other.

```python
plt.figure(figsize=(7, 5))
corr = df[features].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.savefig('heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved: heatmap.png')
```

## Step 8: Key Findings and Insights

Based on the exploration above, here are the main takeaways:

1. **No missing values** – the dataset is clean and ready for modeling.
2. **Balanced classes** – 50 samples per species, no class imbalance to worry about.
3. **Petal features are the strongest separators** – petal length and petal width clearly distinguish all 3 species, especially *Iris setosa* which is completely separate from the other two.
4. **Sepal features overlap** – sepal length and sepal width alone cannot reliably separate *virginica* from *versicolor*.
5. **High correlation** – petal length and petal width are strongly correlated (r ≈ 0.96), meaning they carry similar information.
6. **Outliers** – sepal width has a few mild outliers in the *setosa* class, but nothing that would significantly affect a model.

**Conclusion:** The Iris dataset is well-structured and petal measurements are the most informative features for classification.
